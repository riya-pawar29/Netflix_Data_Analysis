```python
#Average Movie Duration
#We'll first convert the duration into numbers, then calculate:
movies = df[df["type"]=="Movie"].copy()

movies["duration"] = movies["duration"].str.replace(" min","",regex=False)

movies["duration"] = pd.to_numeric(movies["duration"], errors="coerce")

movies = movies.dropna(subset=["duration"])

movies.head()
```

```python
movies["duration"].mean()
```

```python
movies["duration"].max()
```

```python
movies["duration"].min()
```

```python
#Longest Movie
movies.loc[movies["duration"].idxmax(), ["title", "duration"]]
```

```python
#Shortest Movie
movies.loc[movies["duration"].idxmin(), ["title", "duration"]]
```

```python
#Average Movie Duration
movies["duration"].mean()
```

```python
plt.figure(figsize=(6,4))
df["type"].value_counts().plot(kind="bar")
plt.title("Movies vs TV Shows on Netflix")
plt.xlabel("Content Type")
plt.ylabel("Count")
plt.show()
```

##### Insight

Movies are available in greater numbers than TV Shows, showing that Netflix's library is primarily focused on movies.

```python
#Histogram (Movie Duration Distribution)

plt.figure(figsize=(8,5))
plt.hist(movies["duration"], bins=20)
plt.title("Distribution of Movie Duration")
plt.xlabel("Duration (Minutes)")
plt.ylabel("Number of Movies")
plt.show()
```

##### Insight

Most Netflix movies have a duration between 80 and 120 minutes, while very short and very long movies are less common.

```python
#Box Plot (Movie Duration)

plt.figure(figsize=(6,4))
sns.boxplot(x=movies["duration"])
plt.title("Box Plot of Movie Duration")
plt.xlabel("Duration (Minutes)")
plt.show()
```

##### Insight
Shows the spread of movie durations and detects outliers.
The box plot shows that most movies have similar durations, but a few movies are much longer or shorter than the average.    

```python
#Count Plot (Content Type)

plt.figure(figsize=(6,4))
sns.countplot(data=df, x="type")
plt.title("Movies vs TV Shows")
plt.xlabel("Content Type")
plt.ylabel("Count")
plt.show()
```

##### Insight
Counts the number of Movies and TV Shows.
    Netflix has more Movies than TV Shows, indicating that movies dominate the platform's content library.

```python
#Scatter Plot (Release Year vs Movie Duration)

plt.figure(figsize=(8,5))
plt.scatter(movies["release_year"], movies["duration"])
plt.title("Release Year vs Movie Duration")
plt.xlabel("Release Year")
plt.ylabel("Duration (Minutes)")
plt.show()
```

##### Insight
Shows the relationship between release year and movie duration.There is no strong relationship between a movie's release year and its duration. Movies of different lengths have been released across different years.

```python
#Heatmap (Missing Values)

plt.figure(figsize=(10,6))
sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
plt.title("Missing Values Heatmap")
plt.show()
```

##### Insight
Displays the location of missing values in the dataset.The heatmap helps identify missing values across different columns, making it easier to assess the completeness of the dataset.

```python
#Correlation Heatmap

numeric_df = movies[["release_year", "duration"]]

plt.figure(figsize=(5,4))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

##### Insight
* Shows the correlation between numeric variables.
*  The correlation heatmap indicates the relationship between release year and movie duration. The weak correlation suggests that movie duration is largely independent of the release year.
